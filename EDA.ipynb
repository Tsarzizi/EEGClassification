```python
def formatting(data):
    data = np.array(data)
    length = data.shape[0]
    width = data.shape[1]
    temp = np.zeros((62, 200*240))
    if width > 200*240:
        width = 200*240
        temp[:, :width] = data[:, :width]
    else :
        temp[:length, :width] = data
    data = temp
    return data
```

```python
import re

def load_data(data_files_path, save_files_path):
    data_files_name = os.listdir(data_files_path)
    for f in data_files_name:
        match = re.match('^\d.*mat$', f)
        if match is not None:
            f = match.group()
            file_name = osp.join(data_files_path, f)
            print('Loading', file_name)
        
            datas = sio.loadmat(file_name)
            keys = list(datas.keys())
            for i in range(3, len(keys)):
                data = datas[keys[i]]
                data = formatting(data)
                save_name = f[:-4]+'_'+str(i-2)+'.npz'
                save_path = osp.join(save_files_path, save_name)
                np.savetxt(save_path, data)
                print('Saved', save_path)
    print('Done')
load_data('D:\Data\EEG\SEED\\Preprocessed_EEG\\', 'D:\Data\EEG\SEED\\save')
```
